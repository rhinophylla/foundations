## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
#using Matplotlib
columns_of_interest = ["Sample_size", "Median", "Employed", "Full_time", "Unemployment_rate", "Men", "Women"]

plt.figure(figsize=(10,10))
for i in range(1,8):
    plt.subplot(3,3,i)
    plt.hist(majors[columns_of_interest[i-1]])
    plt.gca().set_title(columns_of_interest[i-1])
    plt.show
```

```python
# Using pandas
majors_cols = majors[["Sample_size", "Median", "Employed", "Full_time", "Unemployment_rate", "Men", "Women"]].copy()
majors_cols.hist(figsize=(10,10))
```
